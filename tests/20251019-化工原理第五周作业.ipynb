```python
import numpy as np
import matplotlib.pyplot as plt

# 已知参数
F, x_F = 100, 0.48
D, x_D = 50, 0.95
W, x_W = 50, 0.01
target_stages = 8  # 目标板数

# 计算不同回流比下的理论板数
R = np.linspace(0.1, 10, 100)
stages=[len(McCabeThiele(vle=vle, x_D=x_D, x_W=x_W, x_F=x_F, q=1, R=r,D=D,W=W,F=F).calculate_stages(start_from='top'))-1 for r in R]

# 找到理论板数等于目标板数的回流比（考虑数值计算误差）
tolerance = 1e-6  # 容差
mask = np.isclose(stages, target_stages, atol=tolerance)
intersection_R = R[mask]

# 提取第一个和最后一个交点（若存在）
if len(intersection_R) > 0:
    first_R = intersection_R[0]  # 第一个交点的回流比
    last_R = intersection_R[-1]  # 最后一个交点的回流比
else:
    first_R = last_R = None
    print("未找到交点")

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(R, stages, label='Number of Stages vs Reflux Ratio')
plt.plot(R, [target_stages]*len(R), 'r--', label=f'Target Stages ({target_stages})')

# 标记交点
if first_R is not None and last_R is not None:
    plt.scatter([first_R, last_R], [target_stages, target_stages], 
                color='blue', s=100, zorder=5, label='Intersections')
    # 标注交点坐标
    plt.text(first_R, target_stages+5, f'({first_R:.2f}, {target_stages})', 
             ha='center', color='blue')
    plt.text(last_R, target_stages-5, f'({last_R:.2f}, {target_stages})', 
             ha='center', color='blue')

plt.xlabel("Reflux Ratio (R)")
plt.ylabel("Number of Stages")
plt.title("Effect of Reflux Ratio on Number of Stages")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
